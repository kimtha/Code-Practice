```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("./data/190626-01-E-bottom.csv")
```

```python
df.head()
```

```python
w=np.mean(df["pixel_width"])
h=np.mean(df["pixel_height"])
x=df["position_x"]
y=df["position_y"]
```

```python
print(w)
print(h)
```

```python
fig=plt.figure(figsize=(4,2))
ax=fig.add_subplot(1,1,1)
ax.scatter(x,y,s=2)
ax.set_xlim(0,w)
ax.set_ylim(0,h)
ax.plot(x,y)
```

```python
dfa=pd.read_csv("./data/190626-01-A-bottom.csv")
```

```python
wa=np.mean(dfa["pixel_width"])
ha=np.mean(dfa["pixel_height"])
xa=dfa["position_x"]
ya=dfa["position_y"]
```

```python
fig=plt.figure(figsize=(8,2))
ax=fig.add_subplot(1,2,1)
ax.scatter(xa,ya,s=2)
ax.set_xlim(0,wa)
ax.set_ylim(0,ha)
ax.plot(xa,ya)

dx=fig.add_subplot(1,2,2)
dx.scatter(x,y,s=2)
dx.set_xlim(0,w)
dx.set_ylim(0,h)
dx.plot(x,y)
```
